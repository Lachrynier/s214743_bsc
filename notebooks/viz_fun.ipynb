```python
import numpy as np
import matplotlib.pyplot as plt

# Example: Create a 3D volume with random data
volume = np.random.rand(20, 20, 20)  # Adjust size for more detailed example

def display_slices(volume, slice_nums, axis=0, rows=1, cols=None):
    num_slices = volume.shape[axis]
    
    if isinstance(slice_nums, int):
        # Generate equally spaced slice numbers
        slice_nums = np.linspace(0, num_slices - 1, slice_nums, dtype=int)
    
    total_slices = len(slice_nums)
    
    if cols is None:
        # Calculate columns based on the number of rows and total slices
        cols = (total_slices + rows - 1) // rows  # Ceiling division

    fig, axs = plt.subplots(rows, cols, figsize=(cols * 3, rows * 3), squeeze=False)
    
    for i, slice_num in enumerate(slice_nums):
        if axis == 0:
            slice_data = volume[slice_num, :, :]
        elif axis == 1:
            slice_data = volume[:, slice_num, :]
        else:
            slice_data = volume[:, :, slice_num]
        
        ax = axs[i // cols, i % cols]
        ax.imshow(slice_data, cmap='gray')
        ax.set_title(f'Slice {slice_num}')
        ax.axis('off')
    
    # Hide any extra axes if there are more subplots than slices
    for j in range(i + 1, rows * cols):
        axs[j // cols, j % cols].axis('off')
    
    plt.tight_layout()
    plt.show()

# Example usage:
display_slices(volume, slice_nums=5, axis=0, rows=2)  # 8 slices, 2 rows, columns calculated automatically
```

```python
import numpy as np
import matplotlib.pyplot as plt

def display_slices(volume, axis=0, slices=None, rows=1, cols=None, slice_range=(None, None), fig_scale=1.0):
    num_slices = volume.shape[axis]
    slice_start, slice_end = slice_range if slice_range else (None, None)
    slice_start = 0 if slice_start is None else slice_start
    slice_end = num_slices - 1 if slice_end is None else slice_end

    if slices is None:
        slices = 5
    if isinstance(slices, int):
        # Adjust the range to skip the very first and last points
        step = (slice_end - slice_start) / (slices + 1)
        new_start = slice_start + step
        new_end = slice_end - step
        slices = np.linspace(new_start, new_end, slices, dtype=int)

    total_slices = len(slices)
    if cols is None:
        cols = (total_slices + rows - 1) // rows

    fig, axs = plt.subplots(rows, cols, figsize=(cols * 3 * fig_scale, rows * 3 * fig_scale), squeeze=False)
    for i, slice_num in enumerate(slices):
        ax = axs.flat[i]
        ax.imshow(volume[slice_num, :, :] if axis == 0 else (volume[:, slice_num, :] if axis == 1 else volume[:, :, slice_num]), cmap='gray')
        ax.set_title(f'Slice {slice_num}', fontsize=12 * fig_scale, backgroundcolor='black', color='white', y=1.0)
        ax.axis('off')

    for ax in axs.flat[i+1:]:
        ax.set_visible(False)

    plt.tight_layout(pad=0.4 * fig_scale, h_pad=1.0 * fig_scale, w_pad=1.0 * fig_scale)
    plt.show()

# Example usage
volume = np.random.rand(1000, 20, 20)  # Example volume data
display_slices(volume, axis=0, slices=4, rows=2, fig_scale=1.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def display_slices(volume, slice_info, rows=1, cols=None, fig_scale=1.0, magnify=1.0):
    if cols is None:
        # Calculate columns based on the number of rows and total slices
        cols = (len(slice_info) + rows - 1) // rows  # Ceiling division

    fig, axs = plt.subplots(rows, cols, figsize=(cols * 3 * fig_scale, rows * 3 * fig_scale), squeeze=False)

    for i, (slice_num, axis) in enumerate(slice_info):
        if magnify > 1.0:
            other_axes = [x for x in range(3) if x != axis]
            center = np.array([volume.shape[ax] // 2 for ax in other_axes])
            length = np.array([volume.shape[ax] // magnify for ax in other_axes])
            min_idx = np.round(center - length // 2).astype(int)
            max_idx = np.round(center + length // 2).astype(int)

            if axis == 0:
                slice_data = volume[slice_num, min_idx[0]:max_idx[0], min_idx[1]:max_idx[1]]
            elif axis == 1:
                slice_data = volume[min_idx[0]:max_idx[0], slice_num, min_idx[1]:max_idx[1]]
            else:
                slice_data = volume[min_idx[0]:max_idx[0], min_idx[1]:max_idx[1], slice_num]
        else:
            if axis == 0:
                slice_data = volume[slice_num, :, :]
            elif axis == 1:
                slice_data = volume[:, slice_num, :]
            else:
                slice_data = volume[:, :, slice_num]

        ax = axs.flatten()[i]
        ax.imshow(slice_data, cmap='gray')
        ax.axis('off')
        # Set subtitle in the upper left corner with a background
        ax.text(0.05, 0.95, f'Axis {axis}, Slice {slice_num}', color='white', ha='left', va='top', 
                fontsize=12 * fig_scale * 0.5, transform=ax.transAxes, 
                bbox=dict(facecolor='black', alpha=0.5, edgecolor='none', boxstyle='round,pad=0.5'))

    # Hide any extra axes if there are more subplots than slices
    for j in range(i + 1, rows * cols):
        axs.flatten()[j].axis('off')

    pad_size = 0.5
    plt.tight_layout(pad=0, h_pad=pad_size * fig_scale, w_pad=pad_size * fig_scale)
    plt.show()

# Example usage
volume = np.random.rand(100, 100, 100)  # Example volume data
slice_info = [(50, 0), (25, 1), (75, 2)]  # List of tuples (slice number, axis)
display_slices(volume, slice_info, rows=1, fig_scale=1.5, magnify=2.0)
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Mock figures with subplots for demonstration (replace with your actual figures)
fig1, ax1 = plt.subplots(2, 1)  # Suppose fig1 has 2 vertical subplots
ax1[0].plot([1, 2, 3], [1, 4, 9], color='red')
ax1[1].plot([1, 2, 3], [1, 0, -1], color='red')

fig2, ax2 = plt.subplots(3, 1)  # Suppose fig2 has 3 vertical subplots
ax2[0].plot([1, 2, 3], [1, 4, 9])
ax2[1].plot([1, 2, 3], [5, 2, 5])
ax2[2].plot([1, 2, 3], [1, 4, -100])

fig3, ax3 = plt.subplots(1, 1)  # Suppose fig3 has 1 plot
ax3.plot([1, 2, 3], [1, 4, 100], color='green')

# Creating a new figure to combine them
combined_fig = plt.figure(figsize=(8, 15))  # Adjust the figsize as necessary
gs = GridSpec(6, 1, figure=combined_fig)  # Adjust this based on total subplots needed

# Function to replot from an existing axis to a new axis
def replot_to_new_ax(old_ax, fig, gs, row, col, rowspan, colspan):
    new_ax = fig.add_subplot(gs[row:row+rowspan, col:col+colspan])
    for line in old_ax.get_lines():
        new_ax.plot(line.get_xdata(), line.get_ydata(), marker=line.get_marker(), color=line.get_color())
    new_ax.set_title(old_ax.get_title())
    new_ax.set_xlabel(old_ax.get_xlabel())
    new_ax.set_ylabel(old_ax.get_ylabel())

# Transfer plots
replot_to_new_ax(ax1[0], combined_fig, gs, 0, 0, 1, 1)
replot_to_new_ax(ax1[1], combined_fig, gs, 1, 0, 1, 1)
replot_to_new_ax(ax2[0], combined_fig, gs, 2, 0, 1, 1)
replot_to_new_ax(ax2[1], combined_fig, gs, 3, 0, 1, 1)
replot_to_new_ax(ax2[2], combined_fig, gs, 4, 0, 1, 1)
replot_to_new_ax(ax3, combined_fig, gs, 5, 0, 1, 1)

# Adjust layout and save
combined_fig.tight_layout()
# plt.savefig('combined_figures.png')
plt.show()
```
